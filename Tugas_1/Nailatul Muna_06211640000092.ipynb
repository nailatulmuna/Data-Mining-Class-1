```python
import pandas as pd
import numpy as np
```

```python
#load Data ke-1
penduduk_kota = pd.read_excel (r'D:/Penduduk.xlsx', sheet_name = 'perkotaan')
penduduk_kota
```

```python
# load data ke-2
penduduk_desa = pd.read_excel (r'D:/Penduduk.xlsx', sheet_name = 'perdesaan')
penduduk_desa
```

```python
# merging data ke-1 dengan data ke-2
penduduk = pd.merge(penduduk_kota, penduduk_desa, on='Provinsi')
penduduk
```

```python
# delete coloumn No
penduduk = penduduk.drop(['No_x','No_y'],axis=1)
penduduk
```

```python
# delete row 33 (indonesia)
penduduk = penduduk.drop([33])
penduduk
```

```python
# insert new coloumn
penduduk['Total Penduduk']=(penduduk['Laki-laki_kota']+penduduk['Perempuan_kota']+penduduk['Laki-laki_desa']+penduduk['Perempuan_desa'])
penduduk
```

```python
# load data ke-2
luas_prov = pd.read_excel (r'D:/Penduduk.xlsx', sheet_name = 'luas')
luas_prov = luas_prov.drop(['No'], axis=1)
luas_prov = luas_prov.drop([33])
luas_prov
```

```python
# merging Total Penduduk dengan Luas_prov
data = pd.merge(penduduk,luas_prov,on='Provinsi')
data = data.drop(['Laki-laki_kota','Perempuan_kota','Laki-laki_desa','Perempuan_desa'], axis=1)
data
```

```python
# penjumlahan baris
a = data.sum(axis=0, skipna=True)
a
```

```python
# membuat baris baru
b = pd.Series(['Indonesia',a[1],a[2]], index = ['Provinsi','Total Penduduk','Luas'])
data = data.append(b,ignore_index=True)
data
```

```python
# menghitung kepadatan penduduk
data['Kepadatan'] = (data['Total Penduduk']/data['Luas'])
data
```

```python
# interpretasi baris 33 : yaitu Indonesia
x = data['Provinsi']
y = data['Kepadatan']
s = str(x[33])
r = str(y[33])
print('kepadatan penduduk '+ s + ' tahun 2010 sebesar '+ r + ' artinya bahwa secara rata-rata tiap 1 kilometer persegi wilayah di ' 
      + s +' didiami oleh '+ r + ' penduduk.')
```
