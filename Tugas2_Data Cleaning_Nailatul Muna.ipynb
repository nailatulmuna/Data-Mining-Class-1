# Imputation Missing Value

```python
# mendeteksi missing value (NA)
data.isnull().sum()
```

```python
#imputasi missing value data numerik [mean / median]
data['suicides_no'] = data['suicides_no'].fillna((data['suicides_no'].mean())) 
data['suicides/100k pop'] = data['suicides/100k pop'].fillna((data['suicides/100k pop'].mean()))
data['HDI for year'] = data['HDI for year'].fillna((data['HDI for year'].mean()))
data['gdp per capita'] = data['gdp per capita'].fillna((data['gdp per capita'].mean()))
data.isnull().any()
```

# Encode Label

```python
from sklearn.preprocessing import LabelEncoder
```

```python
# membuat koding [Encode] untuk sex, generation, dan age tanpa membuat kolom baru
data['sex'] = LabelEncoder().fit_transform(data['sex'])
data['generation'] = LabelEncoder().fit_transform(data['generation'])
data['age'] = LabelEncoder().fit_transform(data['age'])
data.head()
```

```python
data1 = data.drop(['country-year'], axis=1)
data1
```

# Standardize

```python
from sklearn.preprocessing import StandardScaler
```

```python
data1['suicides_no'] = StandardScaler().fit(data1[['suicides_no']]).transform(data1[['suicides_no']])
data1['population'] = StandardScaler().fit(data1[['population']]).transform(data1[['population']])
data1['suicides/100k pop'] = StandardScaler().fit(data1[['suicides/100k pop']]).transform(data1[['suicides/100k pop']])
data1['HDI for year'] = StandardScaler().fit(data1[['HDI for year']]).transform(data1[['HDI for year']])
data1['gdp per capita'] = StandardScaler().fit(data1[['gdp per capita']]).transform(data1[['gdp per capita']])
data1.head()
```

# Normalize

```python
from sklearn.preprocessing import Normalizer 
```
